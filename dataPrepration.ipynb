```python
import os
import shutil  # it is used to copy the image
import glob
from tqdm import tqdm   # it is used for to show the progress bar
```

```python
Raw_DIR=r'C:\Users\admin\Desktop\driver_do\mrlEyes_2018_01'
for dirpath, dirname, filenames in os.walk(Raw_DIR):
    for i in tqdm([f for f in filenames if f.endswith('.png')]):
        if i.split('_')[4] == '0':
            shutil.copy(dirpath+'/'+i, dst=r'C:\Users\admin\Desktop\driver_do\Prepareddata\close_eyes')
        elif i.split('_')[4] == '1':
            shutil.copy(dirpath+'/'+i, dst=r'C:\Users\admin\Desktop\driver_do\Prepareddata\open_eyes')
                
```
